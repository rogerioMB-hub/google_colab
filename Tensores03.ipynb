# Vector operations

```python
import torch
import matplotlib.pyplot as plt

t_one = torch.tensor([1, 2, 3])
t_two = torch.tensor([4, 5, 6])
```

```python
t_one + 5
```

# Inner product (dot operator)

```python
dot_product = torch.dot(t_one, t_two)
dot_product
```

Produto interno, equivalente ao produto matricial (t_one.t_two').
Observar que, no caso matricial e de forma discreta, t_two deve ser transposto = 1.4 + 2.5 + 3.6 = 4+10+18 = 32.

Dúvidas? https://en.wikipedia.org/wiki/Dot_product

# Linear space

```python
x = torch.linspace(1, 10, 10)
y = torch.exp(x)
y
```

# Matplotlib

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y)
plt.show()
```

Obs: caso não funcione com tensores, dependendo da versão da matplotlib, será preciso converter o tensor para array numpy.

```python
plt.plot(x.numpy(), y.numpy())
plt.show()
#
```

Obs: uma forma de tornar a curva mais suave é aumentando o número de pontos. Assim, passando de 10 para 100 pontos temos que:

```python
x = torch.linspace(0, 10, 100)
y = torch.exp(x)
plt.plot(x, y)       # como funcionou, não vou converter para array numpy
plt.show()
```

It is simple to create another tensor variable to receive data from another function, such as the proposed sin(x), as we can see below.

```python
z = torch.sin(x)
plt.plot(x, z)
plt.show()
```
